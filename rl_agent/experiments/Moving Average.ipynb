```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
results_path = 'pose_easy_6/train_logs_1.txt'
```

```python
res_df = pd.read_csv(results_path)
```

```python
res_df['float_success'] = res_df['success'].apply(lambda x: 1.0 if x else 0.0)
```

```python
aux_s = res_df['float_success'].rolling(window=100).mean().dropna()
plt.plot(aux_s)
plt.ylim((0,1))
plt.yticks(np.arange(0, 1.1, 0.1))
plt.grid()
plt.show()
```

```python
aux_s = res_df['episode_return'].rolling(window=100).mean().dropna()
plt.plot(aux_s)
plt.grid()
```
